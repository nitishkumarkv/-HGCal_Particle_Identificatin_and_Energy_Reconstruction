```python
trim_ahcal_lay = np.array([ 3,  7, 11, 15, 19, 23, 27, 31, 35, 39])
```

```python
def new_rechits(df_):
    
    new_cell_idxs = []
    new_ahc_hit_x = []
    new_ahc_hit_y = []
    new_ahc_hit_z = []
    new_ahc_hit_ene = []
    new_ahc_hit_lay = []
    
    rec_lay = df_.ahc_hitK
    
    for lay in range(1,40):
        
        if lay not in trim_ahcal_lay: continue
    
        rec_x = df_.ahc_hitX[rec_lay==lay]
        rec_y = df_.ahc_hitY[rec_lay==lay]
        rec_z = df_.ahc_hitZ[rec_lay==lay]
        rec_ene = df_.ahc_hitEnergy[rec_lay==lay]
        
        new_cell_idxs = []

        for rec_i in range(rec_x.size):


            xy_array = np.array(np.array([rec_x[rec_i], rec_y[rec_i]]).tolist()).round(1)
            #print(xy_array)


            cell_idx = cell_map[:,2][(cell_map[:,0:2]==xy_array).all(axis=1)][0]
            #print(cell_idx,"\n")


            if cell_idx not in new_cell_idxs:

                new_cell_idxs.append(cell_idx)

                new_ahc_hit_x.append(new_xy_cell[new_xy_cell[:,2]==cell_idx][0][0])
                new_ahc_hit_y.append(new_xy_cell[new_xy_cell[:,2]==cell_idx][0][1])
                new_ahc_hit_z.append(rec_z[rec_i])
                new_ahc_hit_ene.append(rec_ene[rec_i])
                new_ahc_hit_lay.append(lay)

            else: new_ahc_hit_ene[new_cell_idxs.index(cell_idx)] += rec_ene[rec_i]
            
    
    
    return [np.array(new_ahc_hit_x), np.array(new_ahc_hit_y), np.array(new_ahc_hit_z), np.array(new_ahc_hit_ene), np.array(new_ahc_hit_lay)]
            
              
```

```python
new_rechit_values = df_dummy.apply(new_rechits, axis=1)

#df_dummy["new_rechit_x"] = new_rechit_values[0][0]
#df_dummy["new_rechit_y"] = new_rechit_values[0][1]
#df_dummy["new_rechit_z"] = new_rechit_values[0][2]
#df_dummy["new_rechit_energy"] = new_rechit_values[0][3]

#df_dummy["new_cell_idxs"] = new_rechit_values[0][4]

df_dummy.head()
```
